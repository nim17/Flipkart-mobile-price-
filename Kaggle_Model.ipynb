```
import numpy as np
def loadGloveModel(gloveFile):
    print("Loading Glove Model")
    f = open(gloveFile,'r', encoding='utf8')
    model = {}
    for line in f:
        splitLine = line.split(' ')
        word = splitLine[0]
        embedding = np.asarray(splitLine[1:], dtype='float32')
        model[word] = embedding
    print ("Done.",len(model)," words loaded!")
    return model
path=f'{glove_path}/glove.840B.300d.txt'
word2idx=loadGloveModel(path);
```

```
class my_dictionary(dict): 
  
    # __init__ function 
    def __init__(self): 
        self = dict() 
          
    # Function to add key:value 
    def add(self, key, value): 
        self[key] = value 
  
# Main Function 
mapping = my_dictionary() 
wor=list(final_dataframe['word'])
i=0
for w in wor:
    mapping.add(w, i)
    i+=1
mapping
```

```
y_encoded=[]
import numpy as np
for w in y:
    vec=np.zeros(words_to_consider)
    try:
        vec[mapping[w]]=1
    except KeyError:
        vec[words_to_consider-1]=1
    y_encoded.append(vec)
y_encoded=np.array(y_encoded)
np.shape(y_encoded)
```

```
import keras
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding

# train_len is 26 and we want to use 25
# np.shape(y)[1] = 50

def create_model(seq_len):
    model = Sequential()
    model.add(LSTM(512, input_shape=(train_len-1,np.shape(X)[2]),return_sequences=True))
    model.add(LSTM(1024))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(units=np.shape(y_encoded)[1], activation='softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
   
    model.summary()
    
    return model
```

```
seq_len=train_len-1
model = create_model(seq_len)
```
